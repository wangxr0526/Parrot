```python
import os
import pandas as pd
import json
```

```python
source_dataset_path = '../../dataset/source_dataset/'

with open(os.path.join(source_dataset_path, 'USPTO_condition_final', 'pistachio_rxnclass2name.json'), 'r', encoding='utf-8') as f:
    rxn_class2name = json.load(f)

def rxn_class2name_function(rxn_class_id):
    rxn_class_id = str(rxn_class_id)
    rxn_superclass, subclass = rxn_class_id.split('.')
    if rxn_class_id not in rxn_class2name:
        return rxn_class2name[rxn_superclass]
    else:
        return rxn_class2name[rxn_class_id]


uspto_condition_df = pd.read_csv(os.path.join(source_dataset_path, 'USPTO_condition_final', 'USPTO_condition_pred_category.csv'))
uspto_condition_df['rxn_class_name'] = uspto_condition_df['rxn_category'].apply(lambda x:rxn_class2name_function(x))
uspto_condition_df
```

```python
org_uspto_condition_with_mapped_rxn = pd.read_csv(os.path.join(source_dataset_path, 'uspto_rxn_condition_remapped_and_reassign_condition_role_rm_duplicate_rm_excess.csv'))
org_uspto_condition_with_mapped_rxn = org_uspto_condition_with_mapped_rxn[['canonical_rxn', 'remapped_rxn', 'confidence']]
org_uspto_condition_with_mapped_rxn = org_uspto_condition_with_mapped_rxn.sort_values(by='confidence', ascending=False)
org_uspto_condition_with_mapped_rxn = org_uspto_condition_with_mapped_rxn.drop_duplicates(subset=['canonical_rxn'], keep='first')
org_uspto_condition_with_mapped_rxn = org_uspto_condition_with_mapped_rxn.loc[org_uspto_condition_with_mapped_rxn['canonical_rxn'].isin(uspto_condition_df['canonical_rxn'])]

```

```python
uspto_condition_df = pd.merge(uspto_condition_df, org_uspto_condition_with_mapped_rxn, on=['canonical_rxn'])
```

```python
uspto_condition_df.to_csv(os.path.join(source_dataset_path, 'USPTO_condition_final', 'USPTO_condition_pred_category_maped_rxn.csv'), index=False)
uspto_condition_df
```
